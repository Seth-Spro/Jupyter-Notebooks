```python
avg_rating = ramen_data.groupby('Country').agg('mean')
avg_rating.plot.bar()
```

```python
total_rating = ramen_data.groupby('Country').agg('count')
total_rating.Stars.plot.bar()
```

```python
max_rating = ramen_data.groupby('Country').agg('max')
max_rating.Stars.plot.bar()
```
