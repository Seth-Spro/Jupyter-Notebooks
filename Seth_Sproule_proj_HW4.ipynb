```python
import pandas as pd
import seaborn as sns
```

```python
ramen_url = 'C:\\Users\\Seith Sproule\\Documents\\UNI\\ramen-ratings.csv'
ramen_data = pd.read_csv(ramen_url)
```

```python
ramen_data.head(5)
```

```python
#Q2
sns.barplot(data=ramen_data, x="Style", y='Stars')
```

```python
#Q3
sns.catplot(ramen_data, kind='bar', x='Style',y='Stars')
```

```python
#Q4
a =sns.catplot(ramen_data, kind='bar', x='Style',y='Stars')
a.set(title='Rating by Ramen Type',xlabel='Ramen Type')
```

```python
#Q5
a =sns.catplot(ramen_data.query('Country in ("Taiwan","USA","Japan","India")'), kind='bar', x='Style',y='Stars',
               col='Country',col_wrap=2)
Country = ramen_data['Country'].drop_duplicates().tolist()
for index, ax in enumerate(a.axes.flat):
    ax.set(title = Country[index] + ' Rating by Ramen Type')
```
