## Ex - Scanner

```python
#Q1
scanner = pd.read_csv(r'C:\Users\Seith Sproule\Downloads\scanner_data.csv')
```

```python
#Q2
scanner.info()
```

```python
scanner.head(5)
```

```python
scanner['Date'] = pd.to_datetime(scanner['Date'], format='%d/%m/%Y')
scanner.head()
```

```python
columns_to_drop = ['Unnamed: 0', 'SKU']
scanner.drop(columns_to_drop, axis=1, inplace=True)
```

```python
scanner.head(5)
```

```python
scanner_g = scanner.groupby(['Date', 'Customer_ID', 'SKU_Category'])[['Quantity', 'Sales_Amount']].sum()
scanner_g.head(5)
```

```python
#Q6 #I used ChatGPT for this question because I kept getting an error every time I tried to simplify the code.
#But I undertsand this code.
grouped_data_reset = scanner_g.reset_index()
grouped_data_reset['Date'] = pd.to_datetime(grouped_data_reset['Date'], format='%d/%m/%Y')
grouped_data_reset.set_index('Date', inplace=True)
monthly_resampled_data = grouped_data_reset.resample('M').sum()
monthly_resampled_data
```

```python
#Q7
monthly_resampled_data[['Quantity','Sales_Amount']].plot()
```

```python
#Q8
grouped_data_reset = scanner_g.reset_index()
grouped_data_reset['Date'] = pd.to_datetime(grouped_data_reset['Date'], format='%d/%m/%Y')
grouped_data_reset.set_index('Date', inplace=True)
quaterly_resampled_data = grouped_data_reset.resample('Q').sum()
quaterly_resampled_data
```

```python
#Q9
quaterly_resampled_data[['Quantity','Sales_Amount']].plot()
```

```python
monthly_resampled_data['Running_Total'] = monthly_resampled_data['Sales_Amount'].cumsum()
monthly_resampled_data
```
